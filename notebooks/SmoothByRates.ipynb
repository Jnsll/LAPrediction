```python
import pandas as pd
approx=0
chronicle=0
site_number=2
MYDIR = (
    "output"
    + "/"
    + "Approx"
    + str(approx)
    + "/Chronicle"
    + str(chronicle)
    + "/SiteTest"
    + str(site_number)
    )
df_rates = pd.DataFrame(columns=['Approx', 'Chronicle', 'Test Site', 'SubCatchment', 'Rate', 'H Error Real', 'H Error Predict'])
for rate in [2, 7, 15, 21, 30, 45, 50, 60, 75, 90, 100, 150, 182, 200, 250, 300, 330, 350, 365, 550, 640, 730, 1000, 1500, 2000, 2250, 3000, 3182, 3652]:
    try:
        df_rate = pd.read_csv(MYDIR + "/Prediction_HErrorValues_SubCatch_Chronicle0_Approx0_K27.32_Slope_Elevation_LC_SAR_Area_CV_HVClusters1_Rate" + str(rate) + ".csv", sep=";")
        #print(df_rate)
        df_rates = pd.concat([df_rates, df_rate])
    except:
        continue
print(df_rates)
```
