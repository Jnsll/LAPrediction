```python
df_ref_runs = df_data.loc[df_data['Rate'] == 1.0]
df_ref_runs.describe()
```

```python
median_ref_time = df_ref_runs['Time'].median()
print(median_ref_time)
t_median = hr.Time(str(median_ref_time), default_unit=hr.Time.Unit.SECOND)
print(t_median.to_humanreadable())
```

```python
mean_ref_time = df_ref_runs['Time'].mean()
print(round(mean_ref_time))

print(get_readable_time(round(mean_ref_time)))
```

```python
min_ref_time = df_ref_runs['Time'].min()
print("min:", get_readable_time(str(min_ref_time)))
max_ref_time = df_ref_runs['Time'].max()
print("max:", get_readable_time(str(max_ref_time)))
```

```python
def get_readable_time(time):
    return hr.Time(str(time), default_unit=hr.Time.Unit.SECOND).to_humanreadable()
    
```
